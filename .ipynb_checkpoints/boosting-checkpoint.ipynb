```python
def catboost_trainer(fold, n_splits, data, func='gmean'):
    
    train_data, train_labels = data
    
    skf = StratifiedKFold(n_splits=n_splits, random_state=42, shuffle=True)
    pr = []
    for  i, (train_index, test_index) in tqdm(enumerate(skf.split(train_data, train_labels))):
        tr_data = train_data[train_index]
        tr_labels = train_labels[train_index]
        test_data = Pool(train_data[test_index], train_labels[test_index])

        cat_model = CatBoostClassifier(iterations=15000,
                                   depth=3,
                                   random_seed=2021,
                                   learning_rate=0.001,
                                   loss_function='MultiClassOneVsAll', #''MultiClass',
                                   early_stopping_rounds=20,
                                      )
        cat_model.fit(tr_data, tr_labels,
            eval_set=test_data,
            use_best_model=True,
            verbose=False,
            plot=False)
        
        cat_model.save_model(f'catboost_models/{fold}_catboost_model_{i}')
        

#catboost_trainer(1, 7, (train_features, train_labels), func='gmean')
```

```python
def catboost_predict(features, boosting_models=os.listdir('catboost_models')):
    preds = []
    for model_name in boosting_models:
        cat_model = CatBoostClassifier().load_model(f'catboost_models/{model_name}')
        preds_proba = cat_model.predict_proba(features)
        preds += [preds_proba]
    return preds
```

```python
def predict_feats(data):
    boosting_models = os.listdir('catboost_models')
    models_list = np.array(os.listdir('final_models'))
    preds = []
    for fold in [1]:
        filt = [(re.search(f'_{fold}_', model) is not None) for model in models_list] 
        fold_models_list = models_list[filt]
        features = get_features(data, fold_models_list, n_tta=4, func='gmean', BS=16,
                 NUM_CORES=NUM_CORES, img_size=528, device=None)
        fold_boosting_models = [model for model in boosting_models if model[0] == str(fold)]
        preds = gmean(catboost_predict(features, fold_boosting_models), axis=0)
        preds += preds
    return preds
```

```python
def show_examples(images):
    examples = len(images)
    width = int(examples ** .5)
    height = examples // width
    _indexes = [(i, j) for i in range(height) for j in range(width)]
    
    f, ax = plt.subplots(height, width, figsize=(16, 16))
    for (i, j), (title, img) in zip(_indexes, images):
        ax[i, j].imshow(img)
        ax[i, j].set_title(title)
    f.tight_layout()

def read_random_images(df: pd.DataFrame, size: int=4):
    data = np.random.choice(df.image_id, size)
    result = []
    for d in data:
    

        #title = '_'.join([str(i) for i in train[train.image_id == name].values[0]])
        title = train[train.image_id == d].label.values#[0]#[3]
        _image = get_img(d)
        result.append((title, _image))
    
    return result

def get_img(path):
    im_bgr = cv2.imread(f'{str(ROOT)}/train/{path}')
    im_rgb = im_bgr[:, :, ::-1]
    #print(im_rgb)
    return im_rgb
```
